```python
import pandas as pd
df=pd.read_csv("/content/tested.csv")
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Survived'], inplace=True)
```

```python
df['Deck'] = df['Cabin'].str[0]
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.drop_duplicates(subset="PassengerId",inplace=True)
```

```python
df.info()
```

```python
df.head(50)
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked', 'Deck'], drop_first=True)
```

```python
df.to_csv("cleaned data.csv",index=False)
```

- Loaded the dataset using Pandas and explored its structure using `head()` and `info()`.
- Identified missing values and cleaned them using appropriate methods such as drop, mean/median, and mode.
- Removed duplicate records to maintain data consistency.
- Converted data types and created new columns to improve data usability.
- Saved the cleaned dataset as a CSV file for further analysis and reporting.
